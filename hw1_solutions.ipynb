#### 5. [10 points] Why should we avoid evaluating machine learning models using the same data that was used to train the model?

***Answer***

<font color="white">

When we evaluate a model using the same data that was used to train it, we are testing how effective the model is for making predictions when presented with points in the training set. Instead, we typically aim to test how well the model will make decisions when presented with new data (points that it has not seen before). These two values can differ significantly. For example, the nearest neighbor algorithm will often make perfect predictions for every point in the training data (if there are no two points with the same input features but different labels), regardless of how accurate the nearest neighbor algorithm would be for making predictions for new (previously unseen) data.

#### 6. [10 points] Propose a new (not discussed in class) evaluation metric for regression.

Give an equation for the new metric, and explain what it does. Does it have potential benefits for some problems? Does it have potential drawbacks?

***Answer***

<font color="white">

There are many possible answers. One example would be a metric that places additional weight on larger errors by raising errors to the fourth power rather than squaring them. We might call this the *mean quartic error* (MQE):

$$\operatorname{MQE}=\frac{1}{n}\sum_{i=1}^n (y_i-\hat y_i)^4.$$

This would be effective for problems where larger errors are far more significant than smaller errors, while it would be inappropriate for problems where smaller errors are nearly as significant as larger errors.

**Note**: Both of the example answers here are not metrics that are commonly used - they were made up in response to this question.

Another example would be a metric that penalizes only errors in one direction. For example, the *mean squared underprediction* (MSU):

$$\operatorname{MSU}=\frac{1}{n}\sum_{i=1}^n \mathbf 1_{(y_i > \hat y_i)}(y_i-\hat y_i)^2,$$

where $\mathbf 1_{(A)}$ is equal to $1$ if $A$ is true and $0$ otherwise. Here the term $\mathbf 1_{(y_i > \hat y_i)}$ multiplies the error by zero (removing it) if the prediction is an over-prediction or correct, and multiplies it by one (leaving it unchanged) if it is an under-prediction.

This second example is useful in cases where overpredictions are not a concern, but underpredictions are. For example, this could be useful in the earlier landslide example, since overpredictions could result in homes being built too far from a slope (a small cost), while underpredictions could result in homes being built too close to a slope (a potential tragedy).

For this problem, we recommend referencing the course notebook "2.1 Pandas and Datasets.ipynb", which provides an example of how the correlation matrix can be displayed using seaborn. You may reference this code when constructing your initial answer.

***Answer***

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data set
kick = fetch_openml(name='kick', version=1, as_frame=True, parser='auto')
df = kick.data
df['label'] = kick.target

# Select the first five numerical columns, and the label
columns = ['PurchDate', 'VehYear', 'VehicleAge', 'VehOdo', 'MMRAcquisitionAuctionAveragePrice', 'label']

# Compute the correlation matrix
corr_matrix = df[columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
# In the line below, fmt='.2e' means to use scientific notation, since otherwise the numbers overflow the cells.
sns.heatmap(corr_matrix, annot=True, fmt='.2e', cmap='coolwarm')
plt.title('Correlation Matrix of First Five Numerical Features and the Label')
plt.show()
```
